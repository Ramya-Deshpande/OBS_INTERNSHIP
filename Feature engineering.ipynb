```python
threshold=0.3
columns_to_drop=df.columns[df.isnull().mean()>threshold]
df.drop(columns=columns_to_drop,inplace=True)
df
```

```python
numerical_columns=df.select_dtypes(include=['int64','float64']).columns
df[numerical_columns]=df[numerical_columns].apply(lambda x: x.fillna(x.median()))
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
df[categorical_columns]=df[categorical_columns].apply(lambda x: x.fillna('Unknown'))
```

```python
df[['Value','Wage','Release Clause']]=df[['Value','Wage','Release Clause']].replace('[€]','',regex=True)
df
```

```python
def convert_currency(value):
    if isinstance(value,str):
        value=value.replace('M','000000').replace('K','000')
        if value.strip().lower()=='unknown':
            return None
        return float(value)
    return value

df[['Value','Wage','Release Clause']]=df[['Value','Wage','Release Clause']].apply(lambda x:x.apply(convert_currency) )
df
```

```python
def height_conversion(value):
    feet,inch=value.split("'")
    cm=(int(feet)*30.48)+(int(inch)*2.54)
    return cm
df['Height']=df['Height'].apply(height_conversion)
df
```

```python
def weight_conversion(value):
    value=value.replace('lbs',"")
    kg=float(value)*0.453592
    return kg
df["Weight"]=df['Weight'].apply(weight_conversion)
df
```

```python
def outliers(numerical_columns,df):
    o={}
    for column in numerical_columns.columns:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-1.5*IQR
        upper_bound=Q3+1.5*IQR
        o[column]=df[(df[column]<lower_bound)|(df[column]>upper_bound)][column]
    return o
outlier_dictionary=outliers(numerical_columns,df)
for column,outliers in outlier_dictionary.items():
    print(column,":",outliers)
```

```python
import matplotlib.pyplot as plt
nc=numerical_columns.columns.tolist()
df_cleaned=df[nc].dropna(axis=1)
plt.figure(figsize=(10, 6)) 
plt.boxplot([df_cleaned[col].dropna() for col in df_cleaned.columns], labels=df_cleaned.columns, vert=False)
```

```python
def remove_outliers(df):
    numerical_columns = df.select_dtypes(include=['number'])  # Select only numeric columns
    for column in numerical_columns.columns:
        if column in df.columns:  # Ensure column exists
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]  

    return df

df_final = remove_outliers(df_cleaned)
df_final
```
